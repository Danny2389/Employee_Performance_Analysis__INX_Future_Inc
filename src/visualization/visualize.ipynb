```python
# Importing relevant libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
# Importing data from the dataset
Emp = pd.read_excel(r'E:\Projects\DATASCIENCE & AI\Internship Projects\Employee-Performance-Analysis--INX-Future-Inc\data\processed\INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xlsx')
```

```python
# To vizualize all the Features in this Dataset
import matplotlib.pyplot as plt
Emp.plot()
```

```python
import seaborn as sns
# to analyse the education background
sns.set(rc={'figure.figsize':(7,7)})
sns.countplot(x="EducationBackground",data=Emp)
```

```python
# to analyse Gender
print("Number of employees in the dataset:" + str(len(Emp.index)))
# analyse the data
sns.countplot(x="PerformanceRating",hue="Gender",data=Emp)
```

```python
# to see agedistribution
Emp["Age"].plot.hist()
```

```python
# to see agedistribution
Emp["YearsSinceLastPromotion"].plot.hist(figsize=(7,5))
```

```python
#Marital Status
print("Number of employees in the dataset:" + str(len(Emp.index)))
# analyse the data
sns.countplot(x="PerformanceRating",hue="MaritalStatus",data=Emp)
```

```python
# to analyse Emp department
print("Number of department in the dataset:" + str(len(Emp.index)))
# analyse the data
sns.countplot(x="PerformanceRating", hue="EmpDepartment", data=Emp)
```

```python

plt.figure(figsize=(8, 5))
sns.histplot(data=Emp, x='Age', bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=Emp, x='EmpDepartment', y='PerformanceRating')
plt.title("Performance Rating by Department")
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(12, 8))
corr = Emp.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
